## Race

```python
race = ['RaceAsian',
 'RaceBlack',
 'RaceOther',
 'RaceWhite']
```

```python
def get_race(data):
    if data['RaceWhite'] == 1:
        return 'White'
    elif data['RaceBlack'] == 1:
        return 'Black'
    elif data['RaceAsian'] == 1:
        return 'Asian'
    else:
        return 'Other'
    
```

```python
data['Race'] = data[race].apply(get_race, axis=1)
```

```python
sns.countplot(x='Race',data=data)
```

```python
plt.figure(figsize=(12,8))
sns.histplot(x='time',data=data, hue='Race', kde=True)
```

```python
data.corr()[race].loc[['time','death']]
```

```python
sns.heatmap(data.corr()[race].loc[['time','death']], annot=True)
```

## Region

```python
region = ['RegionAsia',
 'RegionEastEuro',
 'RegionNorthAmer',
 'RegionSouthAmer',
 'RegionWestEuro']
```

```python
def get_region(data):
    if data['RegionAsia'] == 1:
        return 'Asia'
    elif data['RegionEastEuro'] == 1:
        return 'EastEuro'
    elif data['RegionNorthAmer'] == 1:
        return 'NorthAmer'
    elif data['RegionSouthAmer'] == 1:
        return 'SouthAmer'
    else: 
        return 'WestEuro'
```

```python
data['Region'] = data[region].apply(get_region, axis=1)
```

```python
sns.countplot(x='Region',data=data)
```

```python
plt.figure(figsize=(12,8))
sns.histplot(x='time',data=data, hue='Region', kde=True)
```

```python
plt.figure(figsize=(12,8))
sns.histplot(x='time',data=data[data['Region'] == 'SouthAmer'], kde=True)
```
